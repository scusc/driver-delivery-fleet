```python
df=pd.read_csv('E:\\driver_data.csv')
```

```python
df.dropna()
```

```python
import matplotlib.pyplot as plt
from sklearn import cluster
from sklearn.cluster import KMeans
```

```python
k = cluster.KMeans(n_clusters=4).fit(df)
centroids = k.cluster_centers_
print(centroids)
plt.scatter(df['Distance_Feature'],df['Speeding_Feature'],c= k.labels_.astype(float),alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='blue')
plt.title("driver delivery fleet")
plt.xlabel("distance feature")
plt.ylabel("speeding feature")
plt.show()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
range_n_clusters = list (range(2,10))
for n_clusters in range_n_clusters:
    clusterer = KMeans (n_clusters=n_clusters).fit(df)
    preds = clusterer.predict(df)
    centers = clusterer.cluster_centers_
    score = silhouette_score (df, preds, metric='euclidean')
    model = SilhouetteVisualizer(KMeans(n_clusters))
    model.fit(df)
    model.show()
    print ("For n_clusters = {}, silhouette score is {}".format(n_clusters, score))
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(df, method  = "ward"))
plt.title('Dendrogram')
plt.xlabel('samples')
plt.ylabel('driver delivery fleets')
plt.show()
```

```python
for n_clusters in range_n_clusters:
    ac = AgglomerativeClustering(n_clusters) 
    plt.figure(figsize =(6, 6)) 
    plt.scatter(df['Distance_Feature'], df['Speeding_Feature'],  
    c = ac.fit_predict(df), cmap ='rainbow')
    score = silhouette_score(df, ac.fit_predict(df), metric='euclidean')
    plt.show() 
    print ("For n_clusters = {}, silhouette score is {}".format(n_clusters, score))
```
